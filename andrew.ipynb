```python
df2 = pd.read_csv('artists_enriched.csv', sep='\t')
df2
```

```python
df2_sub = df2[['artist_id', 'name', 'main_genre', 'image_url']]
df2_sub.drop_duplicates(subset=['artist_id'], inplace=True)
df2_sub
```

```python
# replace all - values with None
df2_sub.replace('-', None, inplace=True)
df2_sub
```

```python
df2.columns
```

```python
df3 = pd.read_csv('songs_popularity.csv', sep='\t')
df3
```

```python
artist_year_scores = df2.groupby(['artist_id','name', 'year'])['year_end_score'].sum().reset_index()
artist_year_scores = artist_year_scores.sort_values(by = 'year')
artist_year_scores = artist_year_scores[artist_year_scores["name"] != "Various Artists"]

artist_year_scores
```

```python
# Drop duplicates to avoid merging ambiguities
artist_info = df2_sub.drop_duplicates(subset='artist_id')[['artist_id', 'main_genre', 'image_url']]
artist_year_scores_merged = artist_year_scores.merge(artist_info, on='artist_id', how='left')
artist_year_scores_merged
```

```python
# Give me a dataframe. Each year include the top 10 artists by year_end_score
top_artists_by_year = artist_year_scores_merged.groupby('year').apply(
    lambda x: x.nlargest(7, 'year_end_score')
).reset_index(drop=True)
top_artists_by_year
```

```python
top_artists_by_year['main_genre'].unique()
```

```python
# download artist_year_scores.csv
artist_year_scores_merged.to_csv('artist_year_scores.csv', sep=',', index=False)
```
